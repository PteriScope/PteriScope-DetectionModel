```python
import cv2
import os
import csv
import numpy as np
```

```python
# Ruta del directorio del dataset
dataset_dir = r'\UPC\Tesis\DATASET-FINAL'

# Ruta del archivo CSV de salida
output_csv_path = '/UPC/Tesis/Repositories/pteriscope-detection_model/dataset.csv'
```

```python
# Abrir el archivo CSV para escritura
with open(output_csv_path, mode='w', newline='') as file:
    writer = csv.writer(file)

    # Escribir el encabezado del CSV
    header = ['label'] + [f'pixel{i}' for i in range(256*256*3)]  # Asumiendo imágenes de 256x256x3
    writer.writerow(header)

    # Recorrer todas las carpetas y archivos en el directorio del dataset
    for idx, label_folder in enumerate(os.listdir(dataset_dir)):
        label_folder_path = os.path.join(dataset_dir, label_folder)

        # Verificar si es una carpeta (una clase/etiqueta)
        if os.path.isdir(label_folder_path):
            label = idx  # La etiqueta es el nombre de la carpeta

            # Recorrer todas las imágenes en la carpeta
            for image_file in os.listdir(label_folder_path):
                image_file_path = os.path.join(label_folder_path, image_file)

                # Leer y procesar la imagen
                img = cv2.imread(image_file_path)

                # Verificar si la imagen se ha cargado correctamente
                if img is None:
                    print(f"Error: Unable to load image at path: {image_file_path}")
                    continue

                img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convertir a RGB
                img_flatten = img_rgb.flatten()  # Aplanar la imagen

                # Escribir la etiqueta y los píxeles en el CSV
                writer.writerow([label] + img_flatten.tolist())
```
